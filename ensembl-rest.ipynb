```python
server = "https://rest.ensembl.org"
# ext = "/homology/id/ENSMUSG00000001506?"
ext = "/homology/id/ENSMUSG00000001506?sequence=cdna;target_taxon=10090;target_species=cow;type=paralogues"
 
r = requests.get(server+ext, headers={ "Content-Type" : "application/json"})
 
if not r.ok:
    r.raise_for_status()
    sys.exit()

decoded = r.json()
print(repr(decoded))
```
